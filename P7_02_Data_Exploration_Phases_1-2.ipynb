### Affichage des pourcentages de Nan

```python
for i in range(df.shape[1]):
    pct_nan=df.iloc[:,i].isnull().sum()/len(df)
    col_name=list(df.columns)[i]
    print('{} : {}'.format(col_name, pct_nan))
```

```python
def print_select_pct_nan(pct_tsh):
    counter=0
    for i in range(df.shape[1]):
        pct_nan=df.iloc[:,i].isnull().sum()/len(df)
        col_name=list(df.columns)[i]
        if pct_nan<=pct_tsh:
            print('{} : {}'.format(col_name, pct_nan))
            counter+=1
    print(counter)
```

```python
print_select_pct_nan(0.05)
```

```python
print_select_pct_nan(0.1)
```

```python
df.iloc[:,30].dtype
```

```python
def print_select_pct_nan_num(pct_tsh):
    counter=0
    for i in range(df.shape[1]):
        pct_nan=df.iloc[:,i].isnull().sum()/len(df)
        col_name=list(df.columns)[i]
        if pct_nan<=pct_tsh and df.iloc[:,i].dtype!='object':
            print('{} : {}'.format(col_name, pct_nan))
            counter+=1
    print(counter)
```

```python
print_select_pct_nan_num(0.05)
```

```python
def print_select_pct_nan_to_list(pct_tsh):
    counter=0
    list_col=[]
    for i in range(df.shape[1]):
        pct_nan=df.iloc[:,i].isnull().sum()/len(df)
        col_name=list(df.columns)[i]
        if pct_nan<=pct_tsh:
            list_col.append(col_name)
            counter+=1
    print(list_col)
```

```python
print_select_pct_nan_to_list(0.05)
```
