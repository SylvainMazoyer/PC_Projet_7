### Affichage des pourcentages de Nan

```python
for i in range(df.shape[1]):
    pct_nan=df.iloc[:,i].isnull().sum()/len(df)
    col_name=list(df.columns)[i]
    print('{} : {}'.format(col_name, pct_nan))
```

```python
df.dtypes.value_counts()
```

```python
df.select_dtypes('object').apply(pd.Series.nunique, axis = 0)
```

```python
le = LabelEncoder()
le_count = 0

# Iterate through the columns
for col in df:
    if df[col].dtype == 'object':
        # If 2 or fewer unique categories
        if len(list(df[col].unique())) <= 2:
            # Train on the training data
            le.fit(df[col])
            # Transform both training and testing data
            df[col] = le.transform(df[col])
            
            # Keep track of how many columns were label encoded
            le_count += 1
            
print('%d columns were label encoded.' % le_count)
```

```python
df = pd.get_dummies(df)

print('Training Features shape: ', df.shape)
```

```python
path_transformed='/home/emma_sylvain/Formation_OpenClassrooms_DataScientist/Projet_7/transformed//'
#df.to_csv(path_transformed+'df_train3.csv', index=False)
```

```python
df_10pc=df.iloc[0:int(len(df)/10),:]
old_names = df_10pc.columns.tolist()
new_names = []
for column_name in old_names:
    new_column_name=column_name.replace(' ','_')
    new_column_name=new_column_name.replace(':','_')
    new_column_name=new_column_name.replace(',','_')
    new_column_name=new_column_name.replace('/','_')
    new_names.append(new_column_name)
new_names
```

```python
df_10pc = df_10pc.rename(columns=dict(zip(old_names, new_names)))
```
